# Quiver plots for ODEs

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

## Dynamics Function

```python
def fcn(t, z):
    return t+z
```

## Compute arrow directions

```python
xs = np.linspace(0, 1, 21)
ys = np.linspace(0, 1, 21)
Y, X = np.meshgrid(xs, ys)
U = np.zeros_like(X)
V = np.zeros_like(Y)
```

```python
for xidx,xx in enumerate(xs):
    for yidx,yy in enumerate(ys):
        u = fcn(xx, yy)
        U[xidx,yidx] = 1.
        V[xidx,yidx] = u
```

## Plot arrows, and solution

```python
y0 = np.array([0.1])
sol = solve_ivp(fcn, (0., 1), y0, max_step=0.01)
```

```python
plt.figure(figsize=[8,8])
Q = plt.quiver(X,Y,U,V,units='xy',color='r')
#plt.plot(tt, sol, 'b')
plt.plot(sol.t, sol.y[0], 'b');
```

# Again, but for a 2D model

```python
def fcn(t, z):
    d = 0.1
    tau = 0.1
    return np.array([-d*z[0]+z[1], -d*z[1]-z[0]])/tau
```

```python
xs = np.linspace(-1, 1, 21)
ys = np.linspace(-1, 1, 21)
Y, X = np.meshgrid(xs, ys)
U = np.zeros_like(X)
V = np.zeros_like(Y)
```

```python
for xidx,xx in enumerate(xs):
    for yidx,yy in enumerate(ys):
        u,v = fcn(0, [xx,yy])
        U[xidx,yidx] = u
        V[xidx,yidx] = v
```

```python
y0 = np.array([0, 0.8])
sol = solve_ivp(fcn, (0., 5), y0, max_step=0.01)
```

```python
plt.figure(figsize=[8,8])
Q = plt.quiver(X,Y,U,V,units='xy',color='r')
#plt.plot(tt, sol, 'b')
plt.plot(sol.y[0], sol.y[1], 'b');
```
