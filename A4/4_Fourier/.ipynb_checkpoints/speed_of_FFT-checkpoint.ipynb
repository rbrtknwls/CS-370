# Speed of FFT

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
sizes2 = [2**n for n in [10, 12, 14, 16, 17]]
sizes3 = [2**n-1 for n in [10, 12, 14, 16, 17]]
```

```python
times2 = []
for s in sizes2:
    f = np.random.rand(s)
    start_time = time.time()
    F = np.fft.fft(f)
    times2.append(time.time()-start_time)
```

```python
times3 = []
for s in sizes3:
    f = np.random.rand(s)
    start_time = time.time()
    F = np.fft.fft(f)
    times3.append(time.time()-start_time)
```

```python
plt.figure(figsize=(10,7))
plt.plot(sizes2, times2, 'ro-');
plt.plot(sizes3, times3, 'go-');
plt.legend([r'$2^n$', r'$2^n - 1$']);
plt.xlabel('N'); plt.ylabel('Clock Time (s)');
```
