# A4-Q5: Audio Enhancing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio
```

```python
# Load the audio recording
Omega, f = scipy.io.wavfile.read('recording.wav')
Audio(f, rate=Omega)
```

```python
# Some useful values
N = len(f)    # total number of samples
L = N / Omega # length of sound clip (in seconds)
t = np.arange(0,N) * L/N  # array of time stamps for samples
```

```python
# Corresponding array of sampled frequencies
omega = np.fft.fftshift(np.arange(-N/2, N/2)) / L
```

```python
plt.plot(t, f)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

## (a)

```python
coeffs = np.fft.fft(f)
```

```python
plt.plot(omega, np.abs(coeffs))
plt.title('Modulus of Fourier Coefficients')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Modulus');
```

## (b)

```python
threshold = 4000
filteredCoeffs = coeffs.copy();
for x in range(0, N):
    if (np.abs(omega[x]) > threshold):
        filteredCoeffs[x] = 0;
```

```python
plt.plot(omega, np.abs(filteredCoeffs))
plt.title('Modulus of Fourier Coefficients (after filter)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Modulus');
```

## (c)

```python
ReconstructedAudio = np.fft.ifft(filteredCoeffs)
```

```python
plt.plot(t, ReconstructedAudio)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

```python
Audio(ReconstructedAudio, rate=Omega)
```
